```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 3  # hz
O = 3 * np.pi
fs = 150  # hz
T = 2.2  # s
N = int(T * fs)
```

```python
x = lambda n: 0.9 * np.sin(2 * np.pi * (n / fs) * np.cos(np.pi * (n / fs)) + O)
```

```python
X = [0] * N
for k in range(N):
    X[k] = sum([
        x(n) * np.exp((-1j * np.pi * k * n) / N)
        for n in range(N)
    ])
X
```
