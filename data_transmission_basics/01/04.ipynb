```python
import math
import matplotlib.pyplot as plt
```

```python
fs = 10000  # hz
T = 4  # s
```

```python
size = int(T * fs)
```

```python
def g(n, H):
    t = n / fs
    return (9 / math.pi ** 2) * sum([math.sin((i * t * math.pi) / 2) * math.sin(20 * n * math.pi * t) for i in range(1, H + 1)])
```

```python
result_3 = [g(i, 3) for i in range(size)]
result_3
```

```python
plt.scatter(list(range(size)), result_3)
```

```python
result_6 = [g(i, 6) for i in range(size)]
result_6
```

```python
plt.scatter(list(range(size)), result_6)
```

```python
result_60 = [g(i, 60) for i in range(size)]
result_60
```

```python
plt.scatter(list(range(size)), result_60)
```
