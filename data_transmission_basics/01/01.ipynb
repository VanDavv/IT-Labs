```python
import math
import matplotlib.pyplot as plt
```

```python
f = 3  # hz
O = 3 * math.pi
fs = 150  # hz
T = 2.2  # s
```

```python
n = int(T * fs)
```

```python
x = lambda n: 0.9 * math.sin(2 * math.pi * (n / fs) * math.cos(math.pi * (n / fs)) + O)
```

```python
result = [x(i) for i in range(n)]
result
```

```python
plt.scatter(list(range(n)), result)
```
